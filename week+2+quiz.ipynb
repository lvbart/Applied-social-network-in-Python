```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook

G3 = nx.Graph()
G3.add_edges_from([
    ('F','A'),
    ('F','B'),
    ('F','C'),
    ('F','D'),
    ('F','E'),
    ('A','B'),
    ('B','C'),
    ('C','D'),
    ('D','E'),
    ('E','A'),
    ('A','D'),
    ('A','C'),
])
nx.draw_networkx(G3)
```

```python
nx.clustering(G3,'F')
```

```python
nx.average_clustering(G3)
```

```python
nx.transitivity(G3)
```
